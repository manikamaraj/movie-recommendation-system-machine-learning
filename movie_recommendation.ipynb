```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.item",sep='|',encoding='latin-1',header=None,names=['movieId', 'title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'] )
```

```python
print(movies)
```

```python
ratings = pd.read_csv('https://files.grouplens.org/datasets/movielens/ml-100k/u.data', sep='\t', names=['userId', 'movieId', 'rating', 'timestamp'])
```

```python
ratings
```

```python
print(ratings)
```

```python
merged_df=pd.merge(movies[["movieId","title"]],ratings,on="movieId")
```

```python
merged_df
```

```python
user_movies_rating=merged_df.pivot_table(index='userId',columns='title',values='rating')
```

```python
user_movies_rating
```

```python
user_movies_rating=user_movies_rating.fillna(0)
```

```python
user_movies_rating
```

```python
user_similarity=cosine_similarity(user_movies_rating)
```

```python
user_similarity
```

```python
user_similarity_df=pd.DataFrame(user_similarity,index=user_movies_rating.index,columns=user_movies_rating.index)
```

```python
user_similarity_df
```

```python
def get_user_recommendation(user_id,top_movies=10):
    #retrieve the similar users by rating
    similar_user=user_similarity_df[user_id].sort_values(ascending=False)[1:11]
    #retrieve the movies rating  by the similar users
    similar_user_movies=user_movies_rating.loc[similar_user.index]
    #retrieve the movies rated by the given user
    user_movies=user_movies_rating.loc[user_id]
    #filter out  the movies that is already rated by the user
    new_movies=similar_user_movies.loc[:,user_movies==0]
    #Calculate the average rating for each unrated movie based on similar users ratings.
    movie_ratings=new_movies.mean().sort_values(ascending=False)
    return movie_ratings.head(top_movies)
```

```python
get_user_recommendation(2)
```
